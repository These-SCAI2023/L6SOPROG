```python
#biblioteques et fonctions
import glob, time, json

def lire_fichier(chemin):
    f = open(chemin)
    chaine = f.read().lower()
    f.close()
    
    return chaine

def decouper_chaine(chaine):
    liste_mots = chaine.split()
    
    return liste_mots

def decopuer_chaine_n(chaine, n):
    
    debut = 0
    n_grams = list()
    
    for i in range(len(chaine)-n+1):
        n_grams.append(chaine[debut:n+i])
        debut += 1
    
    return n_grams

N = 10
APPR = "corpus_multi/*/appr/*"
TEST = "corpus_multi/*/test/*"

#chaine1= "Cet article contient la liste des codes ISO 639 (alpha-2). De nombreuses langues ou familles de langues n'ont pas"

#n_grams = decopuer_chaine_n(chaine1, 3)

#print(n_grams)
```

```python
#Création du dictionnaire de langues

dic_langues = {}

docs = glob.glob(APPR)
i = 1
debut = time.time()

for chemin in docs:
    #print(chemin)
    #print("Document {}/{}".format(i, len(docs)), end="\r", flush=True)
    dossier = chemin.split("/")
    #print(dossier)
    langue = dossier[1]
    #print(langue)    
    if langue not in dic_langues:
        dic_langues[langue] = {}

    chaine = lire_fichier(chemin)    
    #print(chaine)
    mots = decouper_chaine(chaine)
    #mots = decopuer_chaine_n(chaine, 3)
    #print(mots)
    
    for mot in mots:
        if mot not in dic_langues[langue]:
            dic_langues[langue][mot] = 1
        else:
            dic_langues[langue][mot] += 1
            #dic_langues[langue][mot] = dic_langues[langue][mot] + 1
    
    i += 1
    
    #print(dic_langues["sl"])
    #1/0
fin = time.time()
temps = fin - debut
print("\nTemps : {}".format(temps))
```

```python
#print(dic_langues.keys())
print(dic_langues["fr"])
```

```python
#Création des modèles de langue

dic_models = {}

for langue, dic_effectifs in dic_langues.items():
    paires = [[effectif, mot] for mot, effectif in dic_effectifs.items()]
    #print(paires)
    liste_tri = sorted(paires, reverse=True)[:N]
    #print(liste_tri)
    a = [mot for effectif, mot in liste_tri]
    dic_models[langue] = a
    #print(a)
    #1/0
    
```

```python
#print(dic_models.keys())
#print(dic_models["fr"])

print("Mots les plus fréquentes par langue:")
for langue, mots in dic_models.items():
    print(langue, mots)
```

```python
w = open("model.json", "w")
w.write(json.dumps(dic_models, indent = 2))
w.close()
```

```python
#Appliquer les modèles de langue

f = open("model.json", "r")
dic_models = json.load(f)
f.close()

predictions = []
reelles = []

docs = glob.glob(TEST)
print("Détection de langue...")

for chemin in docs:
    
    dic_freq_texte = {}
    
    dossiers = chemin.split("/")
    langue = dossiers[1]
    chaine = lire_fichier(chemin)
    mots = decouper_chaine(chaine)
    #mots = decopuer_chaine_n(chaine, 3)
    for mot in mots:
        if mot not in dic_freq_texte:
            dic_freq_texte[mot] = 1
        else:
            dic_freq_texte[mot] += 1
    
    paires = [[effectif, mot] for mot, effectif in dic_freq_texte.items()]
    liste_tri = sorted(paires, reverse=True)[:N]
    plus_frequents = [mot for effectif, mot in liste_tri]
    #print(plus_frequents)
    #1/0
    
    liste_predictions = []
    
    for langue_ref, model in dic_models.items():
        mots_communs = set(model).intersection(plus_frequents)
        NB_mots_communs = len(mots_communs)
        liste_predictions.append([NB_mots_communs, langue_ref])
        #print(langue_ref, mots_communs)

    prediction = sorted(liste_predictions)[-1][1]
    reelle = langue
    predictions.append(prediction)
    reelles.append(reelle)
    #print("Pred",prediction)
    #print("Reelle", reelle)
    #1/0


```

```python
#Évaluation du système
print(predictions[:20])
print("\n")
print(reelles[:20])

#VP
#FP
#FN
dic_eval = {}

for i in range(len(reelles)):
    prediction = predictions[i]
    reelle =  reelles[i]
    
    if reelle not in dic_eval:
        dic_eval[reelle] = {"VP": 0, "FP":0 , "FN": 0}
        
    if reelle == prediction:
        dic_eval[reelle]["VP"] += 1
    
    if reelle != prediction:
        dic_eval[reelle]["FN"] += 1
        
        if prediction not in dic_eval:
            dic_eval[prediction] = {"VP": 0, "FP":0 , "FN": 0}

        dic_eval[prediction]["FP"] += 1
        

```

```python
for langue, valeurs in dic_eval.items():
    print(langue, valeurs)
```

```python
#Précision, Rappel et F1-mesure

gral_precision = gral_rappel = gral_f1 = 0

print("**Résultats**")
print("Langue\tPrécision\tRappel\t\tF1")
print("-"*50)
for langue, valeurs in dic_eval.items():
    
    if valeurs["VP"] == 0 :
        precision = rappel = f1 = 0
    else:
        precision = valeurs["VP"] / (valeurs["VP"] + valeurs["FP"])
        rappel =    valeurs["VP"] / (valeurs["VP"] + valeurs["FN"])
        f1 = (2 * precision * rappel) / (precision + rappel)

    gral_precision += precision
    gral_rappel += rappel
    gral_f1 += f1
    
    print("{}\t{:.3f}\t\t{:.3f}\t\t{:.3f}".format(langue, precision, rappel, f1))
    
#gral_precision = gral_precision / len(dic_eval.items())
gral_precision /= len(dic_eval.keys())
gral_rappel /= len(dic_eval.keys())
gral_f1 /= len(dic_eval.keys())
print("-"*50)
print("Gral\t{:.3f}\t\t{:.3f}\t\t{:.3f}".format(gral_precision, gral_rappel, gral_f1))
```
